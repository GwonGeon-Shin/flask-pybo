```python
from openpyxl import load_workbook
wb = load_workbook(filename='db_test.xlsx')
ws = wb.active

a = 0
same_count = 0
test_bb=[]
test_bb_01=[]

with app.app_context():
    print('최초 시작시 Question모델 데이터 개수 : ',Question.query.count())
    print('='*15)
    for row in range(1, ws.max_row+1):
        # 엑셀 파일 내에 하나씩 읽어 오는 값을~~
        # 기존의 데이터와 비교해야 겠다.
        print_ques=ws.cell(row,1).value #<---- 엑셀 sheet의 our_ref값
        print('엑셀 파일의 our_ref : ',print_ques)
        
        """
        엑셀파일 내 our_ref 기준값이 db 파일내 없을시 검색에 오류 발생을 예상해서
        try except 적용시킨다.
        """
        try:
            test_bb = Question.query.filter(Question.our_ref.ilike(print_ques)).all()
            test_bb_00 = test_bb[0]
            test_bb_01 = test_bb_00.our_ref #our_ref값만 적용되는 변수 "test_bb_01"
            print('데이터 파일의 해당 id값 : ',test_bb)
            print('데이터 파일의 our_ref : ',test_bb_01)
        except:
            #print("데이터 파일 내 일치 our_ref 없음")
            pass
        
        
        
        if print_ques in test_bb_01:
            '''
            해당 부분 코드는
            our_ref값 기준 엑셀파일과,db파일내 동일 our_ref값이 존재하기에 값을 수정한다.
            위의 "데이터 수정 적용 테스트" 예제를 확인할것
            '''
            test_bb_00.pay_date=ws.cell(row,2).value
            db.session.commit()
            
            same_count = same_count+1
            print('데이터 파일 내 일치 our_ref 존재 : ', same_count)
            print('='*15)
        else:
            '''
            our_ref값 기준 엑셀파일과, db파일내 동일 our_ref값 없을때
            엑셀파일의 내용을 db파일내 작성하는 코드
            '''
            try:
                question = Question(our_ref=ws.cell(row,1).value, pay_date=ws.cell(row,2).value, req_name=ws.cell(row,3).value, cus_name=ws.cell(row,4).value, date=ws.cell(row,5).value, number=ws.cell(row,6).value, q_date=ws.cell(row,7).value, q_d_date=ws.cell(row,8).value, d_d_date=ws.cell(row,9).value, d_d_num=ws.cell(row,10).value, p_name=ws.cell(row,11).value, charge_01=ws.cell(row,12).value, charge_02=ws.cell(row,13).value, p_charge=ws.cell(row,14).value, create_date=datetime.now())
                db.session.add(question)
                db.session.commit()
                print('엑셀 our_ref 정보를 db파일내 생성 하였습니다.')
            except:
                db.session.rollback()
                print('입력에 에러가 발생 하였습니다.')
            finally:
                db.session.close()
            
            same_count = same_count+1
            print('데이터 파일 내 신규 작성', same_count)
            print('='*15)
        
```
